```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:\\Users\\dell\\crimeproject\\crime_data_1.csv')
df.head()
```

```python
df.shape
```

```python
df[df.isnull().any(axis=1)].head()
```

```python
from sklearn import preprocessing
number = preprocessing.LabelEncoder()
df['DISTRICT'] = number.fit_transform(df.DISTRICT)

x = df['DISTRICT']
print(df['DISTRICT'].dtype)
print(df['DISTRICT'])
z = np.matrix(df['DISTRICT'])
print(z)
y = df['MURDER']
```

```python
#data =  df.iloc[:,[1,2]].values
#data
#k = df.iloc[:,2:5].values
#k

data = df.iloc[:,1:3].values
data
```

```python
from sklearn.cluster import KMeans
wcss_list = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(data)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1,20),wcss_list)
plt.title("the elbow method graph")
plt.xlabel('no. of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
plt.scatter(x,y,color =  'red',alpha=0.5
            ,edgecolor='k')
plt.show()
```

```python
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
model = KMeans(n_clusters=15,random_state=0)
kmeans = model.fit_predict(data)
kmeans
print(kmeans)
```

```python
labels = model.predict(data)
centroids = model.cluster_centers_
print(centroids)
print(labels)
```

```python
from sklearn.metrics import silhouette_score
print(silhouette_score(data,kmeans))
```

```python
# Visualising the clusters

plt.scatter(data[kmeans == 0, 0], data[kmeans == 0, 1], s = 100, c = 'red',
            label = 'Cluster 1')
plt.scatter(data[kmeans == 1, 0], data[kmeans == 1, 1], s = 100, c = 'blue',
            label = 'Cluster 2')
plt.scatter(data[kmeans == 2, 0], data[kmeans == 2, 1], s = 100, c = 'green',
            label = 'Cluster 3')
plt.scatter(data[kmeans == 3, 0], data[kmeans == 3, 1], s = 100, c = 'cyan',
            label = 'Cluster 4')
plt.scatter(data[kmeans == 4, 0], data[kmeans == 4, 1], s = 100, c = 'magenta',
            label = 'Cluster 5')




plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],
            s = 300, c = 'yellow', label = 'Centroids')
plt.title('crime')

plt.legend()
plt.show()
```
